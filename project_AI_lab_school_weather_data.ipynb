```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
w1=pd.read_csv('/content/drive/MyDrive/weather_data/weather01.txt',sep= '\s+')
w1
```

```python
w2=pd.read_csv('/content/drive/MyDrive/weather_data/weather02.txt',sep= '\s+')
w3=pd.read_csv('/content/drive/MyDrive/weather_data/weather03.txt',sep= '\s+')
w4=pd.read_csv('/content/drive/MyDrive/weather_data/weather04.txt',sep= '\s+')
w5=pd.read_csv('/content/drive/MyDrive/weather_data/weather05.txt',sep= '\s+')
w6=pd.read_csv('/content/drive/MyDrive/weather_data/weather06.txt',sep= '\s+')
```

```python
w2
```

```python
weather=pd.concat([w1,w2,w3,w4,w5,w6],axis='columns')
weather
```

```python
weather.drop(columns='X', inplace=True)
```

```python
weatherm=weather.melt(id_vars=['year','month','measure'])
weatherm
```

```python
weatherm=weatherm.rename(columns={'variable':'day'})
weatherm
```

```python
weatherm.day=weatherm.day.str.replace('X','').astype(int)
weatherm
```

```python
weatherwnan=weatherm.dropna(subset='value')
weatherwnan
```

```python
weatherwnan['date'] = pd.to_datetime(weatherwnan[['year', 'month', 'day']])
weatherwnan = weatherwnan.drop(columns=['year', 'month', 'day'])
weatherwnan
```

```python
weatherf=weatherwnan.pivot(index='date', columns='measure', values='value')
weatherf
```

```python
# Order columns with 'Temperature' first
temp_cols = [col for col in weatherf.columns if 'Temperature' in col]
other_cols = [col for col in weatherf.columns if 'Temperature' not in col]
weatherf = weatherf[temp_cols + other_cols]

# Rename columns
new_columns = []
for col in weatherf.columns:
  if 'Mean' in col:
    new_col = col.replace('Mean', 'mean')
    if 'Temperature' in col:
      new_col = 'Tmean' + col[len('Mean Temperature'):]
    new_columns.append(new_col)
  elif 'Min' in col:
    new_col = col.replace('Min', 'min')
    if 'Temperature' in col:
      new_col = 'Tmin' + col[len('Min Temperature'):]
    new_columns.append(new_col)
  elif 'Max' in col:
    new_col = col.replace('Max', 'max')
    if 'Temperature' in col:
      new_col = 'Tmax' + col[len('Max Temperature'):]
    new_columns.append(new_col)
  else:
    new_columns.append(col)

weatherf.columns = new_columns
weatherf
```

```python
from sklearn.linear_model import LinearRegression

def predict_next_value(df, column_name, n_values):
    """
    Performs linear regression on the last n_values of a specified column
    in a DataFrame and predicts the next value.

    Args:
        df: The input DataFrame.
        column_name: The name of the column to use for prediction.
        n_values: The number of past values to use for the regression.

    Returns:
        The predicted next value.
        Returns an error message if the column does not exist or if n_values is invalid.
    """
    if column_name not in df.columns:
        return "Error: Column not found in DataFrame."

    if n_values <= 1 or n_values >= len(df):
      return "Error: Invalid number of values for regression"

    # Prepare data for linear regression
    X = np.arange(n_values).reshape(-1, 1)
    y = df[column_name].tail(n_values).values

    # Perform linear regression
    model = LinearRegression()
    model.fit(X, y)

    # Predict the next value
    next_x = n_values
    next_value = model.predict([[next_x]])[0]

    return next_value
```

```python
meanT=predict_next_value(weatherf,'meanT',10)
maxT=predict_next_value(weatherf,'maxT',10)
minT=predict_next_value(weatherf,'minT',10)
predictedval=pd.DataFrame({'maxT':[maxT],'meanT':[meanT],'minT':[minT]})
predictedval
```

```python
plt.figure(figsize=(10, 6))
plt.plot(weatherf.index, weatherf['meanT'].values.astype(float), linestyle='--', label='meanT')
plt.plot(weatherf.index, weatherf['minT'].values.astype(float), linestyle='--', label='minT')
plt.plot(weatherf.index, weatherf['maxT'].values.astype(float), linestyle='--', label='maxT')
plt.xlabel('Date')
plt.ylabel('F')
plt.title('Temperatures')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
